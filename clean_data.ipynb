# Clean the data

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('tables.xlsx')
```

```python
columns_name = []
for name in df:
    if name.find('company*'):
        columns_name.append(name)
columns_name
```

```python
# создаем словарь, в котором будем хранить значения для каждой компании
our_list = {}

# циклом пробегаемся по всем именам столбцов
for name_columns in df.columns:
    # первая колонка не называется именем компании, но ее нужно учесть
    if name_columns != 'Unnamed: 0':
        company_name = name_columns
        # чтобы не перезаписать словарь, проверяем, есть ли уже такой ключ в нем
        if company_name not in our_list.keys():
            # если нет, создаем ключ с названием компании и список, куда будем заносить значения
            our_list[company_name] = []
    # здесь пробегаем по значениям столбца
    for value in df[name_columns]:
        # если встречаем в значении столбца имя компании, записываем ее в переменную
        if value in df.columns:
            company_name = value
            # если имя компании еще не встречалось, создаем ключ и список в словаре
            if company_name not in our_list.keys():
                our_list[company_name] = []
        # если значение не название компании, то преобразовываем в тип int
        # и записываем в список соответствующего ключа
        elif (type(value) != float) & (value not in df.columns):
            a = int(value)
            our_list.get(company_name).append(a)
```

```python
# так как столбцы не одинаковой длины, создаем датафрейм таким образов
clear_df = pd.DataFrame({ key:pd.Series(value) for key, value in our_list.items() })
```
